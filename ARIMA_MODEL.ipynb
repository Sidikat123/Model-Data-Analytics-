```python
from math import sqrt
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import het_breuschpagan
from arch import arch_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
```

```python
# Start the timer
start_time = time.time()

print("Loading and preprocessing data...")
# Load the cleaned data
data = pd.read_csv('cleaned_Market_Data.csv')

# Data preprocessing
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
data.set_index('Date', inplace=True)
data = data.sort_index()

# Calculate returns
data['Returns'] = data['PX_LAST'].pct_change(fill_method=None)
data = data.dropna(subset=['Returns'])

# Ensure there is data left after cleaning
if data.empty:
    raise ValueError("No valid data available after cleaning.")

print("Data loaded and preprocessed.")

# Plot the Returns series
print("Plotting the Returns series...")
plt.figure(figsize=(10, 6))
plt.plot(data['Returns'], color='blue')
plt.title('Return Series')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.show()  # Manually close this plot window to continue

# Split the data
print("Splitting the data into training, validation, and test sets...")
train_size = int(len(data) * 0.70)
val_size = int(len(data) * 0.15)
test_size = len(data) - train_size - val_size

train, val, test = data.iloc[:train_size], data.iloc[train_size:train_size+val_size], data.iloc[train_size+val_size:]

print(f"Training set size: {len(train)}")
print(f"Validation set size: {len(val)}")
print(f"Test set size: {len(test)}")

print("Performing ADF test for stationarity...")
# Perform ADF test for stationarity
result = adfuller(data['Returns'])
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')

print("Plotting ACF and PACF...")
# Plot ACF and PACF
fig, axes = plt.subplots(1, 2, figsize=(16, 4))
fig = sm.graphics.tsa.plot_acf(data['Returns'], lags=40, ax=axes[0])
fig = sm.graphics.tsa.plot_pacf(data['Returns'], lags=40, ax=axes[1])
plt.show()  # Manually close this plot window to continue

print("Starting grid search for best ARIMA parameters...")
# Grid search for best ARIMA parameters
best_aic = np.inf
best_order = (0, 0, 0)
best_mdl = None

pq_rng = range(3)  # Range for p and q (0, 1, 2)
d_rng = range(2)   # Range for d (0, 1)

for i in pq_rng:
    for d in d_rng:
        for j in pq_rng:
            try:
                print(f"Evaluating ARIMA({i},{d},{j})...")
                tmp_mdl = ARIMA(train['Returns'], order=(i, d, j)).fit()
                tmp_aic = tmp_mdl.aic
                print(f"ARIMA({i},{d},{j}) - AIC: {tmp_aic}")
                if tmp_aic < best_aic:
                    best_aic = tmp_aic
                    best_order = (i, d, j)
                    best_mdl = tmp_mdl
            except:
                print(f"ARIMA({i},{d},{j}) failed to fit.")
                continue

print(f'Best ARIMA{best_order} - AIC:{best_aic}')

print("Fitting the best ARIMA model...")
# Fit the model
model = ARIMA(train['Returns'], order=best_order)
model_fit = model.fit()

# Extract model summary
model_summary = model_fit.summary()

print(model_summary)

# Create a DataFrame for the model results
model_results = {
    'Metric': ['p', 'd', 'q', 'AIC', 'BIC', 'Log-Likelihood'],
    'Value': [best_order[0], best_order[1], best_order[2], model_fit.aic, model_fit.bic, model_fit.llf]
}

model_results_df = pd.DataFrame(model_results)

# Display the model results
print("\nARIMA Model Results:")
print(model_results_df)

print("Forecasting...")
# Forecast validation and test periods
forecast_val = model_fit.get_forecast(steps=len(val))
forecast_test = model_fit.get_forecast(steps=len(test))

forecast_val_index = val.index
forecast_test_index = test.index

print("Plotting actual vs forecast...")
# Plot actual vs forecast
plt.figure(figsize=(10, 6))
plt.plot(train.index, train['Returns'], label='Train', color='blue')
plt.plot(val.index, val['Returns'], label='Validation Actual', color='blue')
plt.plot(test.index, test['Returns'], label='Test Actual', color='green')
plt.plot(forecast_val_index, forecast_val.predicted_mean, label='Validation Forecast', color='red')
plt.fill_between(forecast_val_index, forecast_val.conf_int()['lower Returns'], forecast_val.conf_int()['upper Returns'], color='red', alpha=0.2)
plt.plot(forecast_test_index, forecast_test.predicted_mean, label='Test Forecast', color='orange')
plt.fill_between(forecast_test_index, forecast_test.conf_int()['lower Returns'], forecast_test.conf_int()['upper Returns'], color='orange', alpha=0.2)
plt.title('Return Forecast')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend()
plt.show()

print("Evaluating the model...")
# Evaluate the model on validation set
mse_val = mean_squared_error(val['Returns'], forecast_val.predicted_mean)
print(f'Validation MSE: {mse_val}')

# Evaluate the model on test set
mae_test = mean_absolute_error(test['Returns'], forecast_test.predicted_mean)
mse_test = mean_squared_error(test['Returns'], forecast_test.predicted_mean)
rmse_test = sqrt(mse_test)
mape_test = mean_absolute_percentage_error(test['Returns'], forecast_test.predicted_mean)
r2_test = r2_score(test['Returns'], forecast_test.predicted_mean)

# Print the evaluation metrics
print(f'Test MAE: {mae_test}')
print(f'Test MSE: {mse_test}')
print(f'Test RMSE: {rmse_test}')
print(f'Test MAPE: {mape_test}')
print(f'Test R-squared: {r2_test}')

print("Performing residual diagnostics...")
# Residual diagnostics
residuals = model_fit.resid

# Plot residuals
plt.figure(figsize=(10, 6))
plt.plot(residuals, label='Residuals', color='blue')
plt.title('Residuals')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.legend()
plt.show()  # Manually close this plot window to continue

# Plot squared residuals
plt.figure(figsize=(10, 6))
plt.plot(residuals**2, label='Squared Residuals', color='blue')
plt.title('Squared Residuals')
plt.xlabel('Date')
plt.ylabel('Squared Residuals')
plt.legend()
plt.show()  # Manually close this plot window to continue

# Plot ACF and PACF of residuals
fig, axes = plt.subplots(1, 2, figsize=(16, 4))
fig = sm.graphics.tsa.plot_acf(residuals, lags=40, ax=axes[0])
fig = sm.graphics.tsa.plot_pacf(residuals, lags=40, ax=axes[1])
plt.show()  # Manually close this plot window to continue

# Add a constant to exog for Breusch-Pagan test
exog = sm.add_constant(np.arange(len(residuals)))

# Perform the Breusch-Pagan test
print("Performing Breusch-Pagan test...")
bp_test = het_breuschpagan(residuals, exog)
bp_test_results = {
    'Metric': ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value'],
    'Value': bp_test
}

bp_test_df = pd.DataFrame(bp_test_results)

print("\nBreusch-Pagan Test Results:")
print(bp_test_df)

# Perform the ARCH test
print("Performing ARCH test...")
arch_test = arch_model(residuals, vol='ARCH')
arch_test_fitted = arch_test.fit(disp='off')
print(arch_test_fitted.summary())

# QQ plot
sm.qqplot(residuals, line='s')
plt.show()  # Manually close this plot window to continue

# Jarque-Bera test
jb_test = sm.stats.jarque_bera(residuals)
jb_test_results = {
    'Metric': ['Jarque-Bera Test Statistic', 'Jarque-Bera p-value'],
    'Value': [jb_test[0], jb_test[1]]
}

jb_test_df = pd.DataFrame(jb_test_results)

```
